```python
print("Insira o número da opção que você deseja realizar:")
opcao = int(input("1 - Adicionar Novo Cliente \n 2 - Exibir Clientes cadastrados \n 3 - Buscar Cliente \n 4 - Atualizar Informações \n 5 - Excluir Cliente \n 6 - Saindo do Sistem \n"))

match(opcao):
  case 1:
    print("Adicione dados do cliente: ")
  case 2:
    print("Exibindo clientes: ")
  case 3:
    print("Insira o nome do cliente que deseja buscar: ")
  case 4:
    print("Atualize as informações do cliente")
  case 5:
    print("Digite o nome do cliente que deseja excluir: ")
  case 6:
    print("Saindo do sistema")
  case __:
    print("Insira uma opção válida")
```
