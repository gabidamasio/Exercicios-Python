```python
num = int(input("Informe um número possitivo ou aperte 0 para sair "))
acumulador=0

while num!=0:
  if num < 0:
       print("Número inválido!")
  else:
       acumulador = acumulador + num

  print ("A soma dos números inseridos foi: ", acumulador)
  num = int(input("Informe um número possitivo ou aperte 0 para sair"))
```
