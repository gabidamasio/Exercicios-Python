```python
banho = tosa = banhotosa = outros = 0

def menu():
    print("1 - Banho")
    print("2 - Tosa")
    print("3 - Banho e Tosa")
    print("4 - Outros")
    return int(input("Informe o serviço desejado: "))

for x in range(5):
    op = menu()

    match op:
        case 1:
            banho += 1
        case 2:
            tosa += 1
        case 3:
            banhotosa += 1
        case 4:
            outros += 1
        case _:
            print("Opção inválida! Tente novamente.")

print("Quantidades:")
print("Banho:", banho)
print("Tosa:", tosa)
print("Banho e Tosa:", banhotosa)
print("Outros:", outros)
```
