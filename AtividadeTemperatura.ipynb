```python
temp = float(input("Insira a temperatura do ambiente: "))

if (temp >= 25 and temp <= 30):
  print("O clima está agradável! Aproveite seu dia!")

elif (temp >= 31 and temp <= 35):
  print("O clima está quente! Hidrate-se corretamente!")

elif (temp >= 36):
  print("Alerta!!!! Calor Extremo!!!")

else:
  print("Clima frio!")
```
